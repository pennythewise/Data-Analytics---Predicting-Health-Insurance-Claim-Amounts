## Section 2: Business Understanding

### 2.1 Problem Statement

Healthcare insurance providers face significant financial risk when they cannot accurately anticipate the medical costs of their policyholders. High-risk patients with conditions such as smoking, obesity, or diabetes tend to generate disproportionately large claims, yet many insurers price premiums using broad demographic buckets rather than individual risk profiles.

**Problem:** How can we use patient demographic and health data to understand and predict insurance claim amounts — enabling fairer premium pricing and better financial risk management?

**Why it matters:** In 2023, the global health insurance market exceeded USD 2 trillion. Mispriced premiums lead to either insurer losses (underpricing) or customer attrition (overpricing). Data-driven claim analysis directly addresses this gap.

---

## Section 4: Data Preprocessing

### 4.1 Handling Missing Values

```python
df_clean = df.copy()

print('Missing values BEFORE treatment:')
print(df_clean.isnull().sum())
```

```python
# age: numerical — impute with MEDIAN (robust to skew/outliers)
age_median = df_clean['age'].median()
df_clean['age'] = df_clean['age'].fillna(age_median)
print(f'age: 5 missing values imputed with median = {age_median}')

# region: categorical — impute with MODE (most frequent category)
region_mode = df_clean['region'].mode()[0]
df_clean['region'] = df_clean['region'].fillna(region_mode)
print(f'region: 3 missing values imputed with mode = "{region_mode}"')

print('\nMissing values AFTER treatment:')
print(df_clean.isnull().sum())
```

**Justification:**
- `age` (numerical, 5 missing = 0.37%): Median imputation was chosen over mean because the `claim` distribution is right-skewed, and age could be similarly influenced. Median is more robust to skew. Dropping only 5 rows would also be acceptable, but imputation retains all 1,340 records.
- `region` (categorical, 3 missing = 0.22%): Mode imputation replaces missing values with the most frequently occurring region, preserving distributional balance without removing data.

### 4.2 Handling Noise and Outliers

```python
# IQR method to detect outliers
def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower) | (data[col] > upper)]
    return outliers, lower, upper

for col in ['bmi', 'bloodpressure', 'claim']:
    out, lo, hi = detect_outliers_iqr(df_clean, col)
    print(f'{col}: {len(out)} outliers | Valid range: [{lo:.2f}, {hi:.2f}]')
```

```python
# Visualise outliers with boxplots
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for i, col in enumerate(['bmi', 'bloodpressure', 'claim']):
    axes[i].boxplot(df_clean[col].dropna(), vert=True, patch_artist=True,
                    boxprops=dict(facecolor='#4C72B0', alpha=0.6))
    axes[i].set_title(f'Boxplot: {col}')
    axes[i].set_ylabel(col)
plt.suptitle('Outlier Detection (IQR Method)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Treatment: Winsorize (cap) outliers — preserves record count
def winsorize_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data[col] = data[col].clip(lower=lower, upper=upper)
    print(f'{col} winsorized to [{lower:.2f}, {upper:.2f}]')
    return data

for col in ['bmi', 'bloodpressure']:
    df_clean = winsorize_iqr(df_clean, col)

# claim: right-skewed by nature — intentionally retained
print('\nclaim: outliers retained — extreme high claims are genuine medical events, not data errors.')
```

**Justification:**
- **BMI** (9 outliers) and **bloodpressure** (62 outliers): Values outside the IQR fence are capped (Winsorized) to the fence boundaries. This preserves all 1,340 records while reducing distortion in downstream statistical models.
- **Claim** (141 outliers): Intentionally retained. In insurance data, high claim amounts represent legitimate catastrophic medical events. Removing or capping them would introduce systematic bias into prediction models.

### 4.3 Data Transformation

```python
# Min-Max Normalisation — applied to numerical predictors only
# claim is the target variable and is intentionally excluded from scaling
scaler = MinMaxScaler()
scale_cols = ['age', 'bmi', 'bloodpressure', 'children']
df_clean[scale_cols] = scaler.fit_transform(df_clean[scale_cols])
print('Min-Max scaling applied to:', scale_cols)
print('Note: claim (target variable) is excluded from scaling.')
```

```python
# Encoding Categorical Variables

# Binary encoding (Yes/No, male/female → 0/1)
binary_map = {'Yes': 1, 'No': 0, 'male': 1, 'female': 0}
for col in ['gender', 'diabetic', 'smoker']:
    df_clean[col] = df_clean[col].map(binary_map)
    print(f'{col} → binary encoded (0/1)')

# One-Hot Encoding for region (4 nominal categories, no inherent order)
region_dummies = pd.get_dummies(df_clean['region'], prefix='region', dtype=int)
df_clean = df_clean.drop(columns=['region'])
df_clean = pd.concat([df_clean, region_dummies], axis=1)
print('region → one-hot encoded into', list(region_dummies.columns))
```

```python
df_clean.head()
```

```python
df_clean.tail()
```

**Justification:**
- **Binary encoding** is used for `gender`, `diabetic`, and `smoker` as each has exactly two categories with clear binary semantics (Yes/No, male/female).
- **One-Hot Encoding** is used for `region` because it has 4 nominal categories with no inherent order. All 4 dummy columns are retained here. If a linear regression model is used downstream, one column should be dropped at that stage to avoid multicollinearity — this is a modelling-phase decision, not a preprocessing requirement.
- **Min-Max Normalisation** is applied to numerical predictors (`age`, `bmi`, `bloodpressure`, `children`) to bring them to a common [0, 1] scale. This ensures no single feature dominates due to its magnitude, which is important for distance-based and gradient-based algorithms. `claim` is excluded as it is the target variable — scaling the target is not required for predictive modelling.

### 4.4 Data Cleaning

```python
# Check for duplicates
dup_count = df_clean.duplicated().sum()
print(f'Duplicate records: {dup_count}')

# Drop identifier and helper columns — no predictive value
cols_to_drop = ['index', 'PatientID']
df_final = df_clean.drop(columns=cols_to_drop)

print(f'\nDropped columns: {cols_to_drop}')
print(f'\nFinal dataset shape: {df_final.shape}')
print('\nFinal columns:')
print(df_final.columns.tolist())
```

**Justification:**
- Data cleaning was performed as the final step, after imputation, outlier treatment, encoding, and normalisation — ensuring the dataset is fully processed before export.
- **No duplicates** were found (confirmed above), so no rows were removed.
- **`index` and `PatientID`** were dropped as they are row identifiers with no predictive value — retaining them would risk model leakage in future phases.
- All original categorical columns (`gender`, `diabetic`, `smoker`, `region`) were replaced by their encoded counterparts during transformation (Section 4.3); no redundant columns remain.
